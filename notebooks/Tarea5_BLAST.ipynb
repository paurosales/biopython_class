2. Hagan el blast de proteínas ribosómicas de la mosca de la fruta contra las proteínas ribosómicas de levadura.

3. Por cada query de mosca, saca el mejor hit de levadura si tiene un E-value < 1.0e-5.

**¿Qué proteína ribosómica se conserva más entre la mosca de la fruta y la levadura?**

```python
from subprocess import Popen, PIPE, STDOUT

# Set the blast program and arguements as strings
blast_prog= "makeblastdb"
blast_args ='-in ../source/yeast-ribosome.fasta -dbtype prot -title yeastdb -out ../tmp/yeastdb'
# # The Popen instance runs a program
proc = Popen(blast_prog + " " + blast_args,
             stdout = PIPE, stderr = STDOUT, shell = True)

blast_prog= 'blastp'
blast_args ='-query ../source/drosoph-ribosome.fasta -db ../tmp/yeastdb -outfmt "6 qseqid sseqid bitscore evalue" -out ../tmp/results.txt'

proc = Popen(blast_prog + ' ' + blast_args, stdout = PIPE, stderr = STDOUT, shell = True)
```

```python
%%bash
head ../tmp/results.txt
```

```python
query = ''
bestHit = 0
hitsList = []
bestHitFinal = [0,0]

print('Resultados de los queries que estén por debajo del umbral:\n')
with open('../tmp/results.txt', 'r') as file:
    for line in file:
        hit = line.split('\t')
        query = hit[0]
        hitsList = []
        # Compare e-value with the desired threshold
        if float(hit[3]) <= 1.0e-5:
            hitsList.append((hit[1], float(hit[2])))
            # Find maximum score and get its info
            index = hitsList.index(max(hitsList))
            bestHitData = [hitsList[index][0], hitsList[index][1]]
            print('Query: {}\n\tBest hit {}\n\tHit score: {}\n'.format(query, bestHitData[0], bestHitData[1]))
            # Save the info only if its score is greater that the last best to get the absolute maximum    
            if bestHitData[1] > bestHit:
                bestHit = bestHitData[1]
                bestHitFinal[0] = query
                bestHitFinal[1] = bestHitData[0]
                evalue = hit[3]
        else: 
            # Ignore queries >1.0e-5
            pass
```

```python
print('Las proteínas ribosómicas más conservadas con un e-value de {} y un hit-score de {} entre mosca y levadura son:\n\n\t{} para Drosophila melanogaster\n\t{} para Saccharomyces cerevisiae'.format(evalue, bestHit, bestHitFinal[0], bestHitFinal[1]))
```
