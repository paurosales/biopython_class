```python
import Bio.PDB.PDBParser

filename = "../source/1kcw.pdb"
structure_id = "1kcw"
parser = Bio.PDB.PDBParser(QUIET = True)
structure = parser.get_structure(structure_id, filename)
chain_A = structure[0]['A']

cysresidues = []

# Extract all cystein residues
for residue in chain_A:
    if residue.get_resname() == 'CYS':
        cysresidues.append(residue)
        
# Generate all cystein residues comparition
for cys_1 in cysresidues:
    cys1_index = cys_1.get_id()[1]
    for cys_2 in cysresidues:
        cys2_index = cys_2.get_id()[1]
        # Focus on 'Sulfur gamma' atom (SG) to consider distances between linked cysteins
        if (cys_1['SG'] - cys_2['SG']) < 8.0 and cys1_index < cys2_index:
            print("Possible di-sulfide bond:", end=" ")
            print("Cys",cys1_index,"-", end=" ")
            print("Cys",cys2_index, end=" ")
            print(round(cys_1['SG'] - cys_2['SG'],2))
```

```python
#import Bio.PDB.PDBParser

filename = "../source/1ce0.pdb"
structure_id = "1ce0"
parser = Bio.PDB.PDBParser(QUIET = True)
structure = parser.get_structure(structure_id, filename)
chain_A = structure[0]['A']
chain_B = structure[0]['B']

lysresidues_A = []
lysresidues_B = []

# Generate list containing al lysines fron A chain
for residue in chain_A:
    if residue.get_resname() == 'LYS':
        lysresidues_A.append(residue)
        
# Generate list containing al lysines fron B chain
for residue in chain_B:
    if residue.get_resname() == 'LYS':
        lysresidues_B.append(residue)
        
for lys_A in lysresidues_A:
    lys_A_NZ = lys_A['NZ']
    lys_A_NZ_index = lys_A.get_id()[1]
    for lys_B in lysresidues_B:
        lys_B_NZ = lys_B['NZ']
        lys_B_NZ_index = lys_B.get_id()[1]
        #Estoy enteresada en el átomo de SG "sulfur-gamma"
        if (lys_A_NZ - lys_B_NZ) < 12.0:
            print("Possible bond using BS3 ", end = " ")
            print("Lys",lys_A_NZ_index,"-", end = " ")
            print("Lys",lys_B_NZ_index, end = " ")
            print(round(lys_A_NZ - lys_B_NZ,2))

```

```python
from Bio import Entrez
Entrez.email = "prosales@lcg.unam.mx"
from urllib.error import HTTPError 

print("USAGE: Input the outpu fasta file name. Indicate the quantity of items to search and the next fields when indicated:\n\tOrganism: only 1\n\tGene(s): separated by a space\n\tIsolate: if just push Enter")
batch_size = 3

out_file = input("Enter the output's file name:  ")
out_handle = open(out_file, "w")

# Initialice list containing all needed queries
number_search = int(input("\nHow many items to consider search:  "))
all_searchs = []

# Generate list containing correct sintax for each query from inputed data
for i in range(0, number_search):
    search = ""
    flag = 1
    # Data inputs
    organism = input("\nOrganism: ")
    genes = input("Gene(s): ").split()
    isolate = input("Isolate:  ")
    search = "(" + organism + "[Organism])"
    # Split genes with the rigth sintax in case of multiple inputed genes
    if len(genes) == 1:
        search += " AND " + genes[0] + "[Gene Name]"
    elif len(genes) > 1:
        search += " AND ("
        for gene in genes:
            if flag == 1:
                search += gene + "[Gene Name]"
                flag += 1
            elif flag > 1 and flag != len(genes):
                search += " OR " + gene + "[Gene Name]"
                flag += 1
            elif flag == len(genes):
                search += " OR " + gene + "[Gene Name])"
            
    # Add isolate field if inputed
    if isolate != "":
        search += " AND " + isolate + "[Isolate]"
    # Add it to the list
    all_searchs.append(search)
            
# Search queries one by one in Nucleotide database
for search in all_searchs:
    print('\nDownloading data from {}'.format(search))
    search_handle = Entrez.esearch(db = "nucleotide", term = search, usehistory = 'y')
    search_results = Entrez.read(search_handle)
    search_handle.close()
    count = int(search_results["Count"])
    webenv = search_results["WebEnv"]
    query_key = search_results["QueryKey"]
    # Download particular records
    for start in range(0, count, batch_size):
        end = min(count, start+batch_size)
        print("Going to download record %i to %i" % (start+1, end))
        attempt = 1
        while attempt <= 3:
            try:
                fetch_handle = Entrez.efetch(db="nucleotide", rettype="fasta", retmode="text",
                                             retstart=start, retmax=batch_size,
                                             webenv=webenv, query_key=query_key)

                break
            # Error exception
            except HTTPError as err:
                if 500 <= err.code <= 599:
                    print("Received error from server %s" % err)
                    print("Attempt %i of 3" % attempt)
                    attempt += 1
                    time.sleep(15)
                else:
                    raise
                
    data = fetch_handle.read()
    fetch_handle.close()
    out_handle.write(data)
    
print('\nDownload finished, fetched sequences saved at {}'.format(out_file))
out_handle.close()
```
