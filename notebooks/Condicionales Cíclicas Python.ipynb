## Operadores básicos en condicionales
Permiten evaluar condicionesestablecidad, regresan los valores booleanos*True* o *False*, en caso de que la comparación sea correcta o incorrecta respectivamente.
<a id='comparacion'></a>
### De comparación
Operadores de comparación nos permiten evaluar condiciones establecidas.

**Interpretación de los operadores**


```python
# ¿2 es igual a 9?
2 == 9
```

```python
# ¿2 es diferente a 9?
2 != 9
```

```python
# ¿2 es menor a 9?
2 < 9
```

```python
# ¿2 es mayor a 9?
2 > 9
```

```python
# ¿2 es menor o igual a 9?
2 <= 9
```

```python
# ¿2 es mayor o igual a 9?
2 >= 9
```

<a id='logicos'></a>
### Lógicos
Permiten evaluar condiciones en conjunto de acuerdo a una relación de acuerdo al operador lógico utilizado.  
<a id='and'></a>
#### AND
El conjunto de condiciones debe cumplirse para regresar *True* de lo contrario es *False*
Se interpreta como: "Si la condición_1 **y** la condición_2 son verdaderas, entonces la evaluación es verdadera"

**Ejemplo**

```python
9 == 9 and 2>1
```

```python
2 != 9 and 2+2==3
```

```python
2 > 9 and 4==4
```

```python
2 > 9 and 5-1==5
```

<a id='or'></a>
#### OR
Al menos una de las condiciones del conjunto debe cumplirse para regresar *True* de lo contrario es *False*
Se interpreta como: "Si la condición_1 **y/o** la condición_2 son verdaderas, entonces la evaluación es verdadera"

**Ejemplo**

```python
9 == 9 or 2>1
```

```python
2 != 9 or 2+2==3
```

```python
2 > 9 or 4==4
```

```python
2 > 9 or 5-1==5
```

<a id='not'></a>
#### NOT
Revierte la evaluación*True* o *False* de la condición.
Se interpreta como: "Si la condición es verdadera, entonces la evaluación es falsa"

**Ejemplo**

```python
not 9 == 9 
```

```python
not 2 == 9
```

## Condicional cíclica
### *While loop*
Permite ejecutar el bloque de código anidado en el *loop* **mientras** la condición sea *True*, se detiene cuando es *False*. 

**Considera que es necesario...**
- Modificar la variable implicada en la condición para evitar una ejecución infinita.
- Identar el bloque de código (una tabulación respecto al nivel del *while*)
### Ejemplos
#### Sintaxis incorrecta

```python
a =1
while a < 10:
print(a)
a += 2
```

#### Sintaxis correcta

```python
a = 1
while a < 10:
    print(a)
    a += 2
```

## Comandos extra 
Para evitar la formación de *loops* infinitos y facilitar el uso de condicionales como *if, else* e *elif* dentro de concdicionales cíclicas; existen comandos que actúan sobre la iteración del *loop*.

***break***      
Rompe el ciclo.

***continue***   
Se salta a la siguiente iteración del ciclo.

***pass***       
Permite dejar bloques de código dentro de las condicionales, generalmente utilizado en borradores.
