## Ejercicio 1
[Regresar al índice](#indice)

Un ORF inicia con un codón inicial y termina, ya sea con un codón final o al final de la cádena.
Utilizando como input una cadena de ADN, obtener la cadena proteica de mayor longitud que puede ser traducida de cualquiera de sus ORF.
Si más de una cadena protéica de longitud maxima existe, entonces podrás emitir cualquier solución.

````python
from Bio.Seq import Seq
from Bio import SeqIO

def obtain_prot(rna):
    # Generate all posible (considering 3 ORFs)proteins given a singls RNA strand
    prot_vect = []
    prot_size_vect = []
    nt = 0
    flag = True
    # Move through the string by single characters
    while flag:
        # Begin the sequence avoiding last start codon considered
        rna = rna[nt:]
        start_codon_index = rna.find("AUG")
        # If there isn't any start codons left or the sequence is over break the loop
        if start_codon_index == -1 or len(rna) < 3:
            flag = False
        else:
            # Generate the protein according to the start codon till the fist stop codon found
            rna = rna[start_codon_index:]
            prot = rna.translate(to_stop = True)
            prot_vect.append(str(prot))
            prot_size_vect.append(len(prot))
            nt += 3
    return (prot_vect, prot_size_vect) 
       
print('This code estimates the largets protein considering all ORFs of a given DNA sequence\n\nEnter the DNA sequence:')
dna = Seq(input().upper())
# DNA to RNA
fwd_rna = dna.transcribe()
rvrs_rna = fwd_rna.reverse_complement()

# Calculate 3 ORF for each RNA strand translation
fwd_prot, fwd_prot_size = obtain_prot(fwd_rna)
rvrs_prot, rvrs_prot_size = obtain_prot(rvrs_rna)

# Get all protein data together
all_prot = fwd_prot + rvrs_prot
all_prot_sizes = fwd_prot_size + rvrs_prot_size
largest_prot = all_prot[all_prot_sizes.index(max(all_prot_sizes))]
print('\nLargest protein considering all ORFs: \n \t{}\n'.format(largest_prot))
````

   ```

   ```


4. Imprime los codones de cada secuencia en los 6 marcos lectura.

````python
# from Bio import SeqIO

def get_triplets(seq):
    # Generate a string with whole fasta sequence splited in triplets
    seq_str = str(seq)
    divided_seq_vect = [(seq_str[i:i+3]) for i in range(0, len(seq_str), 3)]
    divided_seq = ''
    for i in divided_seq_vect:
        divided_seq += i + ' '
    return (divided_seq)
    
filename = ('../source/seq.nt.fa')
for seq_record in SeqIO.parse(filename, "fasta"):
    print('\n>',seq_record.id)
    divided_seq = get_triplets(seq_record.seq)
    print(divided_seq)
    
````

## Ejercicio 3
[Regresar al índice](#indice)

Obtén el *número* de lecturas cuyo promedio de calidad esté por debajo del umbral dado.

*Input:* Un umbral de calidad, junto con un fastq que los contenga.


````python
# from Bio impor SeqIO

def quality_mean(qualities_vect):
    partial_sum = 0
    for quality in qualities_vect:
        partial_sum += quality
    mean = partial_sum / len(qualities_vect)
    return(mean)
  
print('Fastq file path:')
filename = input()     # used ../source/sample.fastq
print('\nQuality threshold to consider:')
threshold = int(input())
count = 0

# Total reads
for record in SeqIO.parse(filename, 'fastq'):
    count += 1
print('\nTotal of %i reads' % count)

# Save records according to the quality evaluation
low_reads = (
    record
    for record in SeqIO.parse(filename, 'fastq')
    if quality_mean(record.letter_annotations['phred_quality']) < threshold
)
# Count and save low quality records on an output fastq file (extra)
count = SeqIO.write(low_reads, '../tmp/low_quality.fastq', 'fastq')
print('From which {} reads with mean quality lower than {} were saved in ../tmp/low_quality.fastq'.format(count, threshold))
````

Consigue el structure del archivo pdb 4hhb 
+ Imprime las coordenadas 3D del átomo C-alpha del aminoacido con residuo número 20 en chain C y asigna ese residuo a una variable. 
+ Calcula el centro geométrico *(centroid)* del aminoácido mencionado anteriormente e imprímelo (esto asume que las masas de los átomos son iguales)
    
    Para calular el centro geométrico se necesitan sacar las coordenadas de todos los átomos del residuo y sacar el promedio.


````python
from Bio import PDB

def centroid(coords_sum_vect, residue):
    # Calculate residue's centroid assuming same mass for all atoms
    mean = []
    for coord_sum in coords_sum_vect:
        mean.append(coord_sum/len(residue))
    return(mean)

filename = '../source/4hhb.pdb'
structure_id = '4hhb'
parser = PDB.PDBParser(QUIET = True)
structure = parser.get_structure(structure_id, filename)

# Access particular atom
wanted_atom = structure[0]["C"][20]["CA"]
residue = structure[0]["C"][20]
all_atoms_coords = [0, 0, 0]

# Sum all atoms' coords 
for atom in residue:
    all_atoms_coords += atom.get_coord() 

centroid_coords = centroid(all_atoms_coords, residue)

print('Atom 3D coords are:', wanted_atom.get_coord())
print('Residue centroid = ', centroid_coords)                
                
````

````python
from Bio.Blast import NCBIXML

filename = '../source/test.blastout.xml'
e_threshold = 1e-10

handle = open(filename)
blast_records = NCBIXML.parse(handle)

print('Query Name \t Hit ID \t\t e-Value    Hsp Number \t Hit Description \n ----------------------------------------------------------------------------------------------------------------')
for blast_record in blast_records:
    for alignment in blast_record.alignments:
        hsp_num = 0
        for hsp in alignment.hsps:
            hsp_num += 1
            if hsp.expect < e_threshold:
                print(' ', blast_record.query_id, '  ', alignment.hit_id, '  ', hsp.expect, '\t', hsp_num, '  ', alignment.hit_def, '\n')
````
